### Appending to a DataFrame

```python
import pandas as pd
import numpy as np
Collection = {'Album Name':['The Stone Roses', 'Definitely Maybe', 'The Queen is Dead'], 'Band Name':['The Stone Roses', 'Oasis', 'The Smiths'], 'Year':[1989,1994,1986]}
df = pd.DataFrame.from_dict(Collection)

New = {'Album Name':'What\'s the Story', 'Band Name':'Morning Glory'}
real = df.append(New, ignore_index = True) # Adds new row to it
real
```
